```python
text = 'Hey! My name is Hans I am a Research Scientist at Meta. I am broadly interested in disinformation, networks, and natural language processing.'
tokens = text.encode('utf-8')
tokens = list(map(int,tokens))
print('----')
print(text)
print('----')
print("length:", len(text))
print(tokens)
print("length:", len(tokens))
```

```python
def get_stats(ids):
    counts = dict()
    for pair in zip(ids, ids[1:]):
        counts[pair] = counts.get(pair,0)+1
    return counts

stats = get_stats(tokens)
stats = {k: v for k, v in sorted(stats.items(), key=lambda item: item[1], reverse=True)}
print(stats)
```

```python
top_pair = max(stats, key = stats.get)
top_pair
```

```python
def merge(ids, pair, idx):
    new_ids  = []
    i= 0 
    while i < len(ids):
        if i < len(ids)-1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
            new_ids.append(idx)
            i+=2
        else:
            new_ids.append(ids[i])
            i+=1
    return new_ids
#print(merge([5,6,6,7,9,1],(6,7), 99))
print(merge([5,6,6,7,9,1],(6,7), 99))
```
